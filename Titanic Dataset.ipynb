```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)

predictions
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(train_data['Sex'])
le.classes_
train_data['Sex'] = le.transform(train_data['Sex'])
test_data['Sex'] = le.transform(test_data['Sex'])
```

```python
train_data['Cabin'].value_counts()
train_data['Cabin'].fillna('UN', inplace=True)
test_data['Cabin'].fillna('UN', inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(pd.concat([train_data['Cabin'], test_data['Cabin']]))
le.classes_
train_data['Cabin'] = le.transform(train_data['Cabin'])
test_data['Cabin'] = le.transform(test_data['Cabin'])
```

```python
train_data['Embarked'].value_counts()
train_data['Embarked'].fillna('UN', inplace=True)
test_data['Embarked'].fillna('UN', inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(train_data['Embarked'])
le.classes_
train_data['Embarked'] = le.transform(train_data['Embarked'])
test_data['Embarked'] = le.transform(test_data['Embarked'])
```

```python
train_data['Ticket'].value_counts()
train_data['Ticket'].fillna('UN', inplace=True)
test_data['Ticket'].fillna('UN', inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(pd.concat([train_data['Ticket'], test_data['Ticket']]))
le.classes_
train_data['Ticket'] = le.transform(train_data['Ticket'])
test_data['Ticket'] = le.transform(test_data['Ticket'])
```

```python
train_data['Age'].fillna(train_data['Age'].mean(), inplace = True)
test_data['Age'].fillna(train_data['Age'].mean(), inplace = True)
```

```python
train_data['Fare'].fillna(train_data['Fare'].mean(), inplace = True)
test_data['Fare'].fillna(train_data['Fare'].mean(), inplace = True)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
y = train_data["Survived"]

features = ["PassengerId", "Pclass", "Sex", "Age", "SibSp", "Parch", "Ticket", "Fare", "Cabin", "Embarked"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = GradientBoostingClassifier()
#model = LogisticRegression()
#model = GaussianNB()
#model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission_4.csv', index=False)
```
